```python
#importing clustering libraries
import pandas as pd
import numpy as np
import seaborn as sn
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

# agglomerative clustering

```python
crime=pd.read_csv("crime_data.csv")
crime
```

```python
crime1=crime.rename({'Unnamed: 0': 'city_names','UrbanPop':'urban_population'},axis=1)
crime1.tail()
```

```python
crime1.info()
```

```python
crime1[crime1.duplicated()]
```

```python
crime1.boxplot(column=['Murder'])
```

```python
#normalization function to normalize the data 
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return (x)
```

```python
#applying the defined norm_func on the crime dataset 
crime_norm= norm_func(crime1.iloc[:,1:])
```

```python
crime_norm
```

```python
# creating dendogram for normalized data
dendogram = sch.dendrogram(sch.linkage(crime_norm, method='complete'))
```

```python
#creating clusters
hc=AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='complete')
```

```python
# save the clusters for chart
#hc=hierarchical clustering
pred_hc=hc.fit_predict(crime_norm)
pred_hc
```

```python
# inserting clusters into original data set 
#here aggl_cluster= agglomerative clustering
crime1['aggl_cluster']= pred_hc
crime1
```

```python
# groupby by agglomarative in which  3 clusters are cosidered
crime1.iloc[:,1:5].groupby(crime1.aggl_cluster).mean()
```

# kmeans clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(crime_norm)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

#wcss within cluster sum of squares
#random_state=0 to to execute the iteration sequentially
# range(1,11) elbow curve range on x axis
```

```python
#from the above we will select 4 clusters 
model=KMeans(n_clusters=4)
model.fit(crime_norm)
```

```python
model.labels_
```

```python
labels=pd.Series(model.labels_)
crime1['kmeansclust']=labels
crime1
```

```python
# groupby by k means clustering in which 4 clusters are there 
crime1.iloc[:,1:5].groupby(crime1.kmeansclust).mean()
```
